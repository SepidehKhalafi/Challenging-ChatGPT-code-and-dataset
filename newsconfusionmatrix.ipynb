```python
import pandas
import numpy
from google.colab import files
import io
import matplotlib.pyplot
import seaborn
import sklearn.metrics

%matplotlib inline
```

```python
uploaded = files.upload()
news = pandas.read_csv(io.BytesIO(uploaded['numeric_data_1.csv']), sep = ";")
```

```python
news.head()
```

```python
ground_truth = news['Ground Truth']
democratic_view = news['ChatGPT’s response from the perspective of a Conservative (Democrat) viewpoint']
republican_view = news['ChatGPT’s response from the perspective of a Rdaical (Republican) viewpoint']


conf_matrix_democrat = sklearn.metrics.confusion_matrix(ground_truth, democratic_view)
print(conf_matrix_democrat)


conf_matrix_republican = sklearn.metrics.confusion_matrix(ground_truth, republican_view)
print(conf_matrix_republican)
```

```python
def plot_confusion_matrix(conf_matrix, title):
    matplotlib.pyplot.figure(figsize=(7, 4))
    seaborn.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False)
    matplotlib.pyplot.xlabel('Predicted')
    matplotlib.pyplot.ylabel('Actual')
    matplotlib.pyplot.title(title)
    matplotlib.pyplot.show()

# Plot the confusion matrices
plot_confusion_matrix(conf_matrix_democrat, 'Confusion Matrix for Democrat Viewpoint')
plot_confusion_matrix(conf_matrix_republican, 'Confusion Matrix for Republican Viewpoint')
```

```python
TP_democrat = conf_matrix_democrat[1, 1]
TN_democrat = conf_matrix_democrat[0, 0]
FP_democrat = conf_matrix_democrat[0, 1]
FN_democrat = conf_matrix_democrat[1, 0]

TPR_democrat = TP_democrat / (TP_democrat + FN_democrat)
TNR_democrat = TN_democrat / (TN_democrat + FP_democrat)
FPR_democrat = FP_democrat / (FP_democrat + TN_democrat)
FNR_democrat = FN_democrat / (FN_democrat + TP_democrat)
accuracy_democrat = sklearn.metrics.accuracy_score(ground_truth, democratic_view)

print(f"true positive rate for democrat {TPR_democrat}")
print(f"true negative rate for democrat {TNR_democrat}")
print(f"false positive rate for democrat {FPR_democrat}")
print(f"false negative rate fro democrat {FNR_democrat}")
print(f"accuracy of democrat {accuracy_democrat}")
```

```python
TP_republican = conf_matrix_republican[1, 1]
TN_republican = conf_matrix_republican[0, 0]
FP_republican = conf_matrix_republican[0, 1]
FN_republican = conf_matrix_republican[1, 0]

TPR_republican = TP_republican / (TP_republican + FN_republican)
TNR_republican = TN_republican / (TN_republican + FP_republican)
FPR_republican = FP_republican / (FP_republican + TN_republican)
FNR_republican = FN_republican / (FN_republican + TP_republican)
accuracy_republican = sklearn.metrics.accuracy_score(ground_truth, republican_view)

print(f"true positive rate for republican {TPR_republican}")
print(f"true negative rate for republican {TNR_republican}")
print(f"false positive rate for republican {FPR_republican}")
print(f"false ngative rate for republican {FNR_republican}")
print(f"accuracy of republican {accuracy_republican}")
```
